You've been asked to create a graph to go in a presentation that your boss is making. It will show the results of a survey of favourite chocolate bars to determine what will be stocked in the vending machines. It's important to get this right!

The data is provided below, already cleaned and in `chocolates` and `votes` lists (you're welcome). Copy this notebook to your own account. Follow the steps and work out the python code to create the graphs shown below.

```
chocolates = ['Aero Mint', 'Bounty', 'Cadbury Dairy Milk', 'Crunchie', 'Flake', 'Galaxy Bar', 'Kit Kat', 'Mars Bar', 'Snickers', 'Twirl', 'Twix', 'Wispa']
votes = [2, 110, 174, 14, 32, 173, 108, 88, 180, 96, 97, 95]
```

## 1. Setup your enviroment

You'll be using matplotlib and seaborn in this project. To start with use the "classic" style of matplot lib and don't turn seaborn on yet. In the next cell put the required import and setup statements to begin this project.

```

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.style.use('classic')
```

## 2. A simple bar chart

First let's see what we have. Make a simple bar chart (appropriate!) using the default matplot lib settings. It should look something like the one below.

```
chocolates = ['Snickers', 'Mars', 'Twix', 'KitKat', 'Bounty']
votes = [50, 70, 30, 90, 20]
plt.figure(figsize=(8,5))
plt.bar(chocolates, votes)
plt.title('Votes for Chocolate Bars')
plt.xlabel('Chocolate Bars')
plt.ylabel('Votes')
plt.show()
```

Okay so the default options look pretty awful. Let's try using seaborn to see if that fixes it.

Next:
1. turn seaborn on,
1. set the style to ticks
1. use sns.barplot(), and
1. remove the spines.

```
# use seaborn to create the bar chart

sns.set(style='ticks')

plt.figure(figsize=(8,5))
sns.barplot(x=chocolates, y=votes)
sns.despine()  # Remove spines
plt.title('Votes for Chocolate Bars')
plt.xlabel('Chocolate Bars')
plt.ylabel('Votes')
plt.show()
```

Pretty but not especially helpful. The default colours are a little nicer and the general graph looks cleaner but it's still not great.

## 3. Make more readable

It's hard to read the names of the chocolate bars and it takes a lot of work to see which is the most popular. Let's change this to a horizontal bar so there's room to read the text and order the chocolates from most to least popular.

1. Sort the data
1. Make [horizontal bar plot](https://seaborn.pydata.org/examples/horizontal_barplot.html)

```
# Reorder the data by votes and make a horizontal bar chart
sorted_data = sorted(zip(votes, chocolates), reverse=True)
sorted_votes, sorted_chocolates = zip(*sorted_data)

plt.figure(figsize=(8,5))
sns.barplot(x=sorted_votes, y=sorted_chocolates)
plt.title('Votes for Chocolate Bars')
plt.xlabel('Votes')
plt.ylabel('Chocolate Bars')
plt.show()
```

# 4. Some nicer colours and good titles

It's easier to read and to see the most popular options now but the default seaborn palette is a bit too pretty.

1. Choose a more dynamic colour palette
2. Add a title and label to the X-axis
3. Can you highlight the 3 most popular bars by changing their colour?

```

palette = sns.color_palette("Blues_r", len(votes))
highlight = ['#d62728' if i < 3 else palette[i] for i in range(len(votes))]

plt.figure(figsize=(8,5))
sns.barplot(x=sorted_votes, y=sorted_chocolates, palette=highlight)
plt.title('Popularity of Chocolate Bars')
plt.xlabel('Votes')
plt.ylabel('Chocolate Bars')
plt.show()
```

## 5. Finally

Now the graph looks pretty good but it will be used in a presentation so try the seaborn set_context option and save the file to send to your boss.

1. Adjust the context to make the graph larger and bolder
2. Save the figure as a png file.

```
# put it all together and save your figure
# Adjust the context for a larger and bolder graph and save as a PNG
sns.set_context("talk")

plt.figure(figsize=(10,6))
sns.barplot(x=sorted_votes, y=sorted_chocolates, palette=highlight)
plt.title('Popularity of Chocolate Bars')
plt.xlabel('Votes')
plt.ylabel('Chocolate Bars')
plt.tight_layout()

# Save the figure
plt.savefig('chocolate_bars_popularity.png', dpi=300)
plt.show()
```
